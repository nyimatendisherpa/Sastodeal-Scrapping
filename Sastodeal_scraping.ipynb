```python
import requests
from bs4 import BeautifulSoup
url="https://www.sastodeal.com/sd-fast/food-essentials/rice-rice-products.html"
url
```

```python
raw_html=requests.get(url)
raw_html
```

```python
soup=BeautifulSoup(raw_html.text,"html.parser")
# soup
```

```python
raw_soup=soup.find('ol', class_=["products","list","items","product-items"])
raw_soup
```

```python
raw_data=raw_soup.find_all('li')
```

```python
# soup.find('ol')
```

```python
first_row=raw_data[0]
```

```python
first_row.strong.text
```

```python
prices=first_row.find_all('span',attrs={'class':'price'})
```

```python
prices[0].text
```

```python
prices[1].text
```

```python
master=[]
for data in raw_data:
  d={}
  d['name']=data.strong.text
  # d['name']=data.find_all('span',attrs={'class':'price'})
  prices=data.find_all('span',attrs={'class':'price'})
  if(len(prices)==2):
    d['old_price']=prices[0].text
    d['new_price']=prices[1].text
  else:
    d['price']=prices[0].text
  master.append(d)
```
